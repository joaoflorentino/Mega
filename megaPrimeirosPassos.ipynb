<h2 style="color:blue"> Conclusão do filtro NaN acima:</h2>
<p> Em cada coluna do dataframe existem 268 linhas contendo dados missing NaN.<br>
 Desta forma serão retirados do dataframe, pois representam jogos que, de alguma forma, os dados não foram informandos pela Caixa Econômica.<br>
 Como os NaN aparecem em todos as linhas das colunas selecionadas, a retirada é a melhor alternativa para limpar o dataframe da falta de informação.<br></p>
 


```python
#Exclusão dos NaN 
new_df.dropna(inplace=True)
```

```python
new_df.head(10)
```

```python
#verificação Dropna funcionou
new_df.isnull().sum()
```

<h3 style="Color:red"> Remoção dos dados NaN </h3>
<p> Podemos ver que os NaN foram removidos do dataframe e agora podemos continuar com valores para análise preliminar. <br>
Em seguida vamos verificar quantos, dos jagos atuais, apresentaram ganhadores.</p>

```python
# filtrando os sorteios nos quais houveram ganhadores 
# Usando drop() onde a coluna Ganhadores_Sena é igual a zero
indice = new_df.loc[(new_df['Ganhadores_Sena'] == 0)] #Aqui separo os que não ganharam 
df2 = new_df.drop(indice.index) # aqui crio novo dataframe somente com concursos ganhadores.
```

```python
# Mostrando dataframe  com os ganhadores 
df2.head(10)
```

```python
# Salvando novo CSV contendo somente ganhadores sem NaN
df2.to_csv('data/somenteGanadores.csv')
```
