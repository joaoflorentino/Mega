```python
# imports iniciais 
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from datetime import datetime
from IPython.core.display import HTML

%matplotlib inline
```

```python
df = pd.read_csv("data/somaDezenasSorteadas.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head(5)
```

```python
# calculo com a sequencia máxima possível 
df2 = df
sqmax = 55+56+57+58+59+60
sqmin = 1+2+3+4+5+6
md = df['Soma'].mean()
df2['Divsqmax']=df['Soma']/sqmax
df2['Divsqmin']=df['Soma']/sqmin
df2['Divmedia']=df['Soma']/md
df2.head(3)
```

```python
df2.to_csv('data/dadosSomaDivMaxDivmin.csv')
```

```python
# Dezenho de um grafico scaterplot para a soma dos numeros divididos pela sequencia máxima
fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(x="Concurso", y="Divsqmax", label="Divsqmax", data= df2)
sns.scatterplot(x="Concurso", y="Divsqmin", label="Divsqmin", data= df2)
sns.scatterplot(x="Concurso", y="Divmedia", label="Divmedia", data= df2)
ax.set_title('Distribuição Da Soma Divididos pela sequencia máxima', fontsize=18)

plt.legend()
plt.ylabel('Divisões', fontsize= 15)
plt.xlabel('Concuros', fontsize= 15)
plt.grid(axis='y')
plt.show
```

<h2 style="color:red">Conclusão do Gráfico acima:</h2>
<p> Supondo-se que fossem sorteados os maiores valores possíveis (sequencia máxima de valores), a soma resultaria em <b> 345 </b> , desta forma fiz a divisão da soma dos números sorteados pelo valor da soma da sequencia máxima.<br>
    Supondo que esta soma dos valores estivessem mais proxima da sequencia máxima a divisão estaria mais próxima do valor <b> 1 </b>, pois em todos os sorteios os numeros estariam concentrados mais próximo dessa sequencia. Isso se confirma analisando o gráfico acima, pois como podemos ver os pontos em azul que representam a divisão da soma dos numeros sorteados pela soma da sequencia máxima que aparecem mais concentrados próximos ao valor <b> 1 </b>.<br>
    Equanto que os valores da divisão pela sequencia mínima, cuja soma vale <b> 21</b>, ficam bem distante dos pontos azuis e verdes. </p>
