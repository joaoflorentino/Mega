<h2 style = "color:darkgreen"> Estratégia de limpeza NaN</h2>
<p> Neste ponto podemos ver que existem valores NaN diferentes para as colunas Cidade e Estados. Neste caso, primeiro fiz a substituição dos valores NaN para as cidade e o estado e então na seguencia fiz a limpeza dos demais valores NaN dos sorteios. <br> 
Os valores NaN foram substituidos pelas letras <strong>"xx"</stong> que significa para este trabalho sem informação.
</p>

```python
# Troca de valores NaN para cidades e estados por xx que significa sem informação
df3.loc[df3['Cidade'].isnull(),'Cidade'] = 'xx'
df3.loc[df3['UF'].isnull(),'UF'] = 'xx'
```

```python
#verificando o estado dos valores NaN 
df3.isnull().sum()
```

```python
df3.shape
```

```python
# Novo dataframe contendo apenas os valores que nos interessam
df4 = df3.dropna()
df4.head(3)
```

```python
df4.shape
```

```python
#criando novo arquivo csv para usos futuros com a planilha limpa sem NaN LP
df4.to_csv('data/ganhadoresPorEstados-LP.csv')
```

```python
# criando o mapa do Brasil com o folium e o json estados
br_estados = 'data/Brasil_Estados.geojson'
geo_json_data = json.load(open(br_estados))
mapa = folium.Map(width=800, height=600, location=[-15.77972, -47.92972], zoom_start=4)
folium.GeoJson(geo_json_data).add_to(mapa)

folium.GeoJson(geo_json_data, style_function=lambda feature:{'fillcolor':'green',
                                                             'color':'darkred',
                                                             'weigth': 0.2,}).add_to(mapa)
mapa
```

<h2 style = "color:darkblue">Contruindo mapa com informações:</h2>
<p>Com o dataframe dfm contendo as informaçẽos de latitude e longitudes e a frequencia de vitorias de cada estado, os mesmos serão adicionados ao mapa com as devidas marcações.
</p>

```python
# criando o mapa do Brasil com o folium e o json estados e colocando as marcações dos ganhadores 
br_estados = 'data/Brasil_Estados.geojson'
geo_json_data = json.load(open(br_estados))
mapa2 = folium.Map(width=800, height=600, location=[-15.77972, -47.92972],tiles="Stamen Terrain", zoom_start=4)
folium.GeoJson(geo_json_data).add_to(mapa2)

folium.GeoJson(geo_json_data, style_function=lambda feature:{'fillcolor':'green',
                                                             'color':'darkred',
                                                             'weigth': 0.2,}).add_to(mapa2)
lat = dfm['La'].values #aqui lemos os valores da latatitude do dataframe dfm
long = dfm['Lo'].values #aqui lemos os valores da longitude do dataframe dfm
freq = dfm['Frq'].values #aqui lemos os valores da Frequencia de ganhadores do dataframe dfm

for la,lo,frq in zip(lat,long,freq):
    folium.Marker([la,lo], popup=frq, tootip="Clique para mais informações").add_to(mapa2)


mapa2
```
