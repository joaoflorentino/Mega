```python
# imports iniciais 
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from datetime import datetime
from IPython.core.display import HTML

%matplotlib inline
```

```python
df = pd.read_csv("data/dadosSomaDivMaxDivmin.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head(5)
```

<h2 style = "color:red"> Continuação das análises de divisões </h2>

<p> Neste novo notebook vou dar continuidade a análise de divisões procurando analisar a distribuição normal dos valores calculados. <p>

<h2 style = "color:green"> Interpretação da curva normal</h2>
<p>
<img src = "imagens/normal3.png" alt="Nada" style = "float:left;width:500px; height:350px"><br>
</p>

<p> A distribuição normal tem como  uma de suas principais características representar os dados amostrais  de uma variável continua, que, para uma amostra ou população, se encontram distribuidos ao redor da media, conforme gráfico acima. </p>
<p> No gráfico temos que a maioria dos dados ficam a -3 e +3 desvio padrão(representado pro sigma) da média. O desvio padrão é uma medida de dispersão em torno da média populacional ou amostral, de uma variável aleatória contínua. </p>

```python
#usando o seaborn grafico com valores da curva normal 
fig, ax = plt.subplots(figsize=(15,8))
sns.distplot(df.Divmedia)
plt.grid(axis='y')
ax.set_title('Distribuição Normal da divisão pela Média', fontsize=18)
plt.xlabel('Divisão Pela Média', fontsize= 15)
```

<h2 style = "color:red"> Grafico 01 - Normal da Divisão pela média - Análise</h2>

<p> O gráfico, através da distribuição de Gaus, nos mostra que a tendencia de valores igual a <b>1</b> que representa a divisão da soma dos numeros sorteados pelo valor da média em todos os sorteios vem a ser uma tendencia do sistema.</p>

```python
fig.savefig('imagens/curvaNormal01-DivMedia.png')
```

```python
#Comparaçao das distribuições normais pelo máximo e pelo Minimo
fig, ax = plt.subplots(figsize=(15,8))
plt.figure(1) # Criação da Área da figura
#Plotagem do primeiro gráfico
plt.subplot(1,2,1)
sns.distplot(df.Divsqmin)
plt.grid(axis='y')
plt.xlabel('Divisão Pelo Mínimo', fontsize= 12)
#Plotagem do segundo gráfico
plt.subplot(1,2,2)
sns.distplot(df.Divsqmax)
plt.grid(axis='y')
plt.xlabel('Divisão Pelo Máximo', fontsize= 12)
plt.show()
```

<h2 style = "color:red"> Grafico 02: </h2>
<p> Dos gráficos apresentados devemos observar que os histrogramas, dos tres graficos (os dois acima e o grafico 01) tem o mesmo perfil pois os valores diferenciam igualmente pela divisão de um número inteiro (média, máximo ou mínimo), no entanto, as escalas representam o que devemos analisar nestas 3 distribuições. </p> </n>
